Найти самую длинную последовательность упорядоченных чисел в RDD и вывести её.

```python
from pyspark.sql import SparkSession
```

```python
def find_longest_sequence(rdd):

    collected_nums = rdd.map(lambda x: int(x)).collect()
    num_set = set(collected_nums)
    longest_streak = []

    for num in collected_nums:
        if num - 1 not in num_set:
            current_num = num
            current_streak = [current_num]

            while current_num + 1 in num_set:
                current_num += 1
                current_streak.append(current_num)

            if len(current_streak) > len(longest_streak):
                longest_streak = current_streak

    return longest_streak


spark = SparkSession.builder.appName("longest_sequence").getOrCreate()

data = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]
rdd = spark.sparkContext.parallelize(data)

longest_sequence = find_longest_sequence(rdd)
print("Самая длинная последовательность:", longest_sequence)

spark.stop()
```
